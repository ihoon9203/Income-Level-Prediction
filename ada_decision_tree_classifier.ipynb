```python
import pandas as pd
import numpy as np
import matplotlib as mlb

from sklearn import tree
```

```python
df = pd.read_csv('./ilp2021f/train_final.csv')
```

```python
train_X = df[['age','workclass','fnlwgt','education','education.num','marital.status','occupation','relationship','race','sex','capital.gain','capital.loss','hours.per.week','native.country']]
train_y = df[['income>50K']]
```

```python
df_test = pd.read_csv('./ilp2021f/test_final.csv')
test_X = df_test[['age','workclass','fnlwgt','education','education.num','marital.status','occupation','relationship','race','sex','capital.gain','capital.loss','hours.per.week','native.country']]
```

```python
train_X = train_X.replace('?',np.NaN)
test_X = test_X.replace('?',np.NaN)
```

Replacing '?' with most frequent value of each column

```python
train_X = train_X.fillna(df.mode().iloc[0])
test_X = test_X.fillna(df.mode().iloc[0])
```

parse categorized values to integers

```python
train_X_dummies = pd.get_dummies(train_X, columns = train_X.columns)
test_X_dummies = pd.get_dummies(test_X, columns = test_X.columns)
```

```python
final_train, final_test = train_X_dummies.align(test_X_dummies, join= 'left', axis = 1)
```

```python
train_y = train_y.fillna(df.mode().iloc[0])
```

```python
entropy_tree = DecisionTreeClassifier(criterion='entropy')
```

```python
entropy_tree.fit(final_train, train_y)
```
